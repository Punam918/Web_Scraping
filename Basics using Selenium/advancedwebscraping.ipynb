```python
with open('smartprix.html', 'r', encoding='utf-8') as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
soup.prettify()
```

```python
container = soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})  # it is alist 
```

```python
for i in container:
    print(i)
```

```python
import numpy as np
```

```python
for i in container:
   print(i.find('h2').text)
   print(i.find('span',{'class':'price'}).text)
   try:
      print(i.find('div',{'class':'score rank-1-bg'}).find('b').text)
   except:
       print(np.nan)
```

```python
name = []
price = []
rating = []
for i in container:
   name.append(i.find('h2').text)
   price.append(i.find('span',{'class':'price'}).text)
   try:
      rating.append(i.find('div',{'class':'score rank-1-bg'}).find('b').text)
   except:
       rating.append(np.nan)
```

```python
len(name)
```

```python
len(price)
```

```python
len(rating)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Name':name,'Price':price,'Rating':rating})
```
